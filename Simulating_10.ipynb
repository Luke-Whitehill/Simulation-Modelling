```python
import random
from scipy import stats
import matplotlib.pyplot as plt

```

```python
#Question1

transitions = 10000
states = []
for i in range(1,transitions):
    state = 0.5
    proposal = random.uniform(state-1,state+1)
    acceptanceprob = min(1,(stats.gamma.pdf(proposal,2,scale=4)/stats.gamma.pdf(state,2,scale=4)))
    u = random.uniform(state-1,state+1)
    if acceptanceprob > u:
        state = proposal
        states.append(state)
    else:
        state = state

mean = sum(states)/len(states)
print(mean)

plt.hist(states, bins = 50)
plt.show()


```

```python
transitions = 10000
states = []
state = 0.5
for i in range(1,transitions):
    proposal = random.uniform(state-1,state+1)
    paccept = min(1,stats.gamma.pdf(proposal,2,scale=4)/stats.gamma.pdf(state,2,scale=4))
    u = random.uniform(0,1)
    if paccept > u:
        # print("ACCEPT")
        state = proposal
        states.append(state)
    else:
        # print("DENY")
        state = state
mean = sum(states)/len(states)
print("Mean", mean)


normals = []
minimum = min(states)
maximum = max(states)
for j in states:
    norm = (j - minimum)/(maximum-minimum)
    normals.append(norm)
normalmean = sum(normals)/len(normals)
print("Mean of normals", normalmean)
plt.hist(states, bins = 50)
plt.show()
plt.hist(normals, bins = 50)
plt.show()
```
