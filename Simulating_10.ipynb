```python
import random
from scipy import stats
import matplotlib.pyplot as plt

```

```python
transitions = 10000
states = []
state = 0.5
for i in range(1,transitions):
    proposal = random.uniform(state-1,state+1)
    paccept = min(1,stats.gamma.pdf(proposal,4,scale=2)/stats.gamma.pdf(state,4,scale=2))
    u = random.uniform(0,1)
    if paccept > u:
        # print("ACCEPT")
        state = proposal
        states.append(state)
    else:
        # print("DENY")
        state = state
mean = sum(states)/len(states)
print("Mean", mean)

  
normals = []
minimum = min(states)
maximum = max(states)
for j in states:
    norm = (j - minimum)/(maximum-minimum)
    normals.append(norm)
normalmean = sum(normals)/len(normals)
print("Mean of normals", normalmean)
plt.hist(states[2001:], bins = 50)
plt.show()
plt.hist(normals[2001:], bins = 50)
plt.show()
```
