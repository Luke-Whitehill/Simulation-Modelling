```python
 
import random
import numpy as np
import numpy
import math
import matplotlib.pyplot as pyplot
import pandas as pd


```

```python
#Question 1
sims = 25
history = []
h = 0.2 #weeks
rodent_arrivals = 0.1*h
pestcontrol_arrivals = 1/130*h
timesteps = round((10*52*(1/0.1)))
for i in range(sims):
    population = 0
    populationlist = []
    for j in range(timesteps):
        u1 = random.uniform(0,1)
        u2 = random.uniform(0,1)
        if u1 <= rodent_arrivals and u2 >= pestcontrol_arrivals:
            population +=1
            populationlist.append(population)
        elif u2 <= pestcontrol_arrivals:
            population = 0
            populationlist.append(population)
        else:
            population = population
            populationlist.append(population)
    history.append(populationlist)

        
for j in range(0,sims):
    pyplot.plot(history[j])
pyplot.title('Individual Simulations of Rodents')
pyplot.xlabel('Time Steps')
pyplot.ylabel('Rodent Count')
pyplot.show()

df = pd.DataFrame(history)
pyplot.plot(df.mean(axis=0))
pyplot.title('Expected Value Rodents Over 25 Simulations')
pyplot.xlabel('Time Steps')
pyplot.ylabel('Rodent Count')
pyplot.show()



```

```python

delta_t = 0.01
timesteps = round(10 * (1/ delta_t ))

sims = 25
state_history = []
for i in range(sims):
    
    state = 0
    
    mu_s = [False, 1, 3, 2, 1]
    state_history.append([])
    for t in range(timesteps):
        if(random.random() < 10*delta_t):
            if(state < 4):
                state+=1
        if(mu_s[state]):
            if(random.random()<mu_s[state]*delta_t):
                state -=1 
        state_history[i].append(state)
        
        prob_of_1 = [0]*timesteps
for i in range(sims):
    
    for j in range(timesteps):
        
        if(state_history[i][j] == 1):
            
            prob_of_1[j] += 1 / sims

mx_timestep = 0
for i in range(timesteps):
    if(prob_of_1[i] > prob_of_1[mx_timestep]):
        mx_timestep = i
print("The time at which the probability of being in state 1 is the largest:\n", mx_timestep*delta_t, "seconds")

```
