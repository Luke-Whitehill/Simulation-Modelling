```python
#Question 1
breaktimelist = []
unfitcount = 0
iterations = 1000
for i in range(0,iterations):
    templist = []
    u1 = random.uniform(0.5,1.5)
    u2 = random.uniform(0.5,1.5)
    u3 = random.uniform(0.5,1.5)
    u4 = random.uniform(0.5,1.5)
    u5 = random.uniform(0.5,1.5)
    templist.append(u1)
    templist.append(u2)
    templist.append(u3)
    templist.append(u4)
    templist.append(u5)
    templist.sort()
    #print(templist.sort())
    #print(templist)
    breaktime = templist[2]
    # print(breaktime)
    breaktimelist.append(breaktime)
    if breaktime < (50/60):
        unfitcount +=1
print(unfitcount/iterations)

    # print(breaktimelist)
    #print(sum(breaktimelist))
print(sum(breaktimelist)/iterations)
```

```python
#Question 2:
outcomes = []
iterations = 2000
for i in range(0,iterations):
    u1 = random.uniform(0,1)
    outcomes.append(np.sqrt(100*u1))

print(outcomes)
```

```python
print("Expected Value: ",sum(outcomes)/iterations)
print("Variance: ",np.var(outcomes))
```
